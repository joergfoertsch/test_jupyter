```python
%matplotlib inline
from ipywidgets import interactive, fixed, FloatSlider, RadioButtons
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import patches as ptc
from scipy import signal
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
def plot_TP2(w_gr = 100, q = 100):
    fig = plt.figure()
    ax1 = plt.subplot(221)
    ax2 = plt.subplot(223)
    ax3 = plt.subplot(122)
    
    sys=signal.TransferFunction([w_gr], [1,w_gr/q,w_gr*w_gr])
    w1=np.arange(0,10000,1)
    w1, mag, phase = signal.bode(sys, w = w1)
    ax1.semilogx(w1, mag)
    ax1.set_xlabel(r'$i \omega$')
    ax1.set_ylabel(r'Amplitude')
    ax2.semilogx(w1, phase) 
    ax2.set_xlabel(r'$i \omega$')
    ax2.set_ylabel(r'Phase')

    w = w_gr + 0j
    all_q = np.arange(0.001, 1000, 0.001) + 0j
    s_pol_pos_all = -w/(2*all_q) + w/(2*all_q)*np.sqrt(1-4*all_q*all_q)
    s_pol_neg_all = -w/(2*all_q) - w/(2*all_q)*np.sqrt(1-4*all_q*all_q)
    ax3.plot(s_pol_pos_all.real, s_pol_pos_all.imag, color='b')
    ax3.plot(s_pol_neg_all.real, s_pol_neg_all.imag, color='orange')
    
    q = q + 0j
    s_pol_pos = -w/(2*q) + w/(2*q)*np.sqrt(1-4*q*q)
    s_pol_neg = -w/(2*q) - w/(2*q)*np.sqrt(1-4*q*q)
    ax3.plot(s_pol_pos.real, s_pol_pos.imag, marker='o', color='b', label='Pol für gegebenes q')
    ax3.plot(s_pol_neg.real, s_pol_neg.imag, marker='o', color='orange', label='Pol für gegebenes q')
    ax3.set_xlim([-500, 500])
    ax3.set_ylim([-500, 500])
    ax3.set_xlabel(r'$\sigma$')
    ax3.set_ylabel(r'$i \omega$')
    plt.tight_layout()
    plt.grid()
    plt.legend()
    plt.show()
```

```python
def plot_BP(w_gr = 100, q = 100):
    fig = plt.figure()
    ax1 = plt.subplot(221)
    ax2 = plt.subplot(223)
    ax3 = plt.subplot(122)
    
    sys=signal.TransferFunction([0,w_gr/q,0], [1,w_gr/q,w_gr*w_gr])
    w1=np.arange(0,10000,1)
    w1, mag, phase = signal.bode(sys, w = w1)
    ax1.semilogx(w1, mag)
    ax1.set_xlabel(r'$i \omega$')
    ax1.set_ylabel(r'Amplitude')
    ax2.semilogx(w1, phase) 
    ax2.set_xlabel(r'$i \omega$')
    ax2.set_ylabel(r'Phase')

    w = w_gr + 0j
    all_q = np.arange(0.001, 1000, 0.001) + 0j
    s_pol_pos_all = -w/(2*all_q) + w/(2*all_q)*np.sqrt(1-4*all_q*all_q)
    s_pol_neg_all = -w/(2*all_q) - w/(2*all_q)*np.sqrt(1-4*all_q*all_q)
    ax3.plot(s_pol_pos_all.real, s_pol_pos_all.imag, color='b')
    ax3.plot(s_pol_neg_all.real, s_pol_neg_all.imag, color='orange')
    
    q = q + 0j
    s_pol_pos = -w/(2*q) + w/(2*q)*np.sqrt(1-4*q*q)
    s_pol_neg = -w/(2*q) - w/(2*q)*np.sqrt(1-4*q*q)
    ax3.plot(s_pol_pos.real, s_pol_pos.imag, marker='o', color='b', label='Pol für gegebenes q')
    ax3.plot(s_pol_neg.real, s_pol_neg.imag, marker='o', color='orange', label='Pol für gegebenes q')
    ax3.plot(0, 0, 'ro', label='Nullstelle')
    ax3.set_xlim([-500, 500])
    ax3.set_ylim([-500, 500])
    ax3.set_xlabel(r'$\sigma$')
    ax3.set_ylabel(r'$i \omega$')
    plt.tight_layout()
    plt.grid()
    plt.legend()
    plt.show()
```

```python
def plot_Notch(w_pol = 100, w_null = 100, q = 1):
    fig = plt.figure()
    ax1 = plt.subplot(221)
    ax2 = plt.subplot(223)
    ax3 = plt.subplot(122)
    
    sys=signal.TransferFunction([1,0,w_null*w_null], [1,w_pol/q,w_pol*w_pol])
    w1=np.arange(0,10000,1)
    w1, mag, phase = signal.bode(sys, w = w1)
    ax1.semilogx(w1, mag)
    ax1.set_xlabel(r'$i \omega$')
    ax1.set_ylabel(r'Amplitude')
    ax2.semilogx(w1, phase) 
    ax2.set_xlabel(r'$i \omega$')
    ax2.set_ylabel(r'Phase')

    
    w = w_pol + 0j
    all_q = np.arange(0.001, 1000, 0.001) + 0j
    s_pol_pos_all = -w/(2*all_q) + w/(2*all_q)*np.sqrt(1-4*all_q*all_q)
    s_pol_neg_all = -w/(2*all_q) - w/(2*all_q)*np.sqrt(1-4*all_q*all_q)
    ax3.plot(s_pol_pos_all.real, s_pol_pos_all.imag, color='b')
    ax3.plot(s_pol_neg_all.real, s_pol_neg_all.imag, color='orange')
    
    q = q + 0j
    s_pol_pos = -w/(2*q) + w/(2*q)*np.sqrt(1-4*q*q)
    s_pol_neg = -w/(2*q) - w/(2*q)*np.sqrt(1-4*q*q)
    s_null = w_null*1j
    ax3.plot(s_pol_pos.real, s_pol_pos.imag, marker='o', color='b', label='Pol für gegebenes q')
    ax3.plot(s_pol_neg.real, s_pol_neg.imag, marker='o', color='orange', label='Pol für gegebenes q')
    ax3.plot(s_null.real, s_null.imag, 'ro', label='Nullstelle')
    ax3.set_xlim([-500, 500])
    ax3.set_ylim([-500, 500])
    ax3.set_xlabel(r'$\sigma$')
    ax3.set_ylabel(r'$i \omega$')
    plt.tight_layout()
    plt.grid()
    plt.legend()
    plt.show()
```

```python
def plot_Notch_RCL(C=0.01, L=0.01, R=0.01):
    w = 1/np.sqrt(C*L) + 0j
    q = np.sqrt(L/C)/R + 0j
    plot_Notch(w_pol = w, w_null = w, q = q)
```

```python
def plot_HP2(w_gr = 100, q = 100):
    plot_Notch(w_pol = w_gr, w_null = 0, q = q)
```
